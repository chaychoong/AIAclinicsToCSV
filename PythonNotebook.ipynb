```python
import openpyxl

# load workbook
wb = openpyxl.load_workbook('clinics.xlsx')
# load worksheet
ws = wb.active

# Get first column most common
zonesDic = {}

for row in ws.iter_rows():
    if zonesDic.get(row[0].value) == None:
        zonesDic[row[0].value] = 1
    else:
        zonesDic[row[0].value] = zonesDic[row[0].value] + 1
        
sortedList = sorted(zonesDic, key=zonesDic.get, reverse=True)

for item in sortedList[:]: # don't remove as iterating. Create a copy, iterate through that copy
    if item == None or not item.isupper():
        sortedList.remove(item)

print(sortedList)
```

```python
# Create a zone list
zones = ['EAST', 'CENTRAL', 'NORTH-EAST', 'NORTH', 'WEST', 'NORTH-WEST', 'CENTRAL-NORTH', 'CENTRAL-EAST', 'CITY', \
         'CENTRAL-SOUTH', 'SOUTH-WEST', 'CENTRAL-WEST', 'SOUTH']
```

```python
outList = []

for row in ws.iter_rows():
    if row[0].value in zones:
        tmpline = []
        #for cell in row:
        for i in range(len(row)):
            cell = row[i].value
            if i == 5:
                cell = "Singapore " + str(cell).zfill(6) # Append "Singapore " to front of postal code
            tmpline.append(cell)
        outList.append(tmpline)
        
print(outList)
```

```python
import csv

headrow = list(ws.rows)[10]
head = []

for cell in headrow:
    head.append(cell.value)

myFile = open('clinics.csv', 'w', newline='')
with myFile:
    writer = csv.writer(myFile)
    writer.writerow(head)
    writer.writerows(outList)
    
print("Write completed")
```
